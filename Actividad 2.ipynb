```python
import pandas as  pd
import matplotlib.pyplot as plt
```

```python
data = {
    'A単o': [2018, 2019, 2020, 2021, 2022],
    'Ventas': [150, 200, 250, 300, 250]}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['A単o'], df['Ventas'], marker='o')
plt.title ('Ventas por A単os')
plt.xlabel('A単os')
plt.ylabel('Ventas')
plt.grid()
plt.show()
```
